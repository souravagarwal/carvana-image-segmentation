## Start Training Code

First we define a generator to pass tuple of (img, mask) to the training code

```python
def my_image_mask_generator(image_gen, mask_gen):
    train_generator = zip(image_gen, mask_gen)
    for (img, mask) in train_generator:
        yield (img[0], mask[0][:,:,:,:1])  # Required to convert the 3D mask as read by ImageDataGenerator into 1 channel

# img = tuple(image, label) -> We do not need label, hence passing just the img[0]
# mask = tuple(mask, label) -> We just pass mask, mask is read as (H, W, 3), but the
# training script expects number of channels to be 1, hence taking the first channel only
```

```python
SEED = 100

BATCH_SIZE = 8
IMG_HEIGHT = 320
IMG_WIDTH = 480
IMG_CHANNELS = 3
ORIG_HEIGHT = 1280
ORIG_WIDTH = 1918

image_data_generator = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.1  # 10% Validation Set
)



train_image_generator = image_data_generator.flow_from_directory(
                'data/train_folder/',
                batch_size = BATCH_SIZE,
                target_size = (IMG_HEIGHT, IMG_WIDTH),
                seed = SEED, subset='training'
                )

validation_image_generator = image_data_generator.flow_from_directory(
                        'data/train_folder/',
                        batch_size = BATCH_SIZE,
                        target_size = (IMG_HEIGHT, IMG_WIDTH),
                        seed = SEED,
                        subset='validation'
                    )

mask_data_generator = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.1
)


train_mask_generator = mask_data_generator.flow_from_directory(
                'data/train_mask_folder/',
                batch_size = BATCH_SIZE,
                target_size = (ORIG_HEIGHT, ORIG_WIDTH),
                seed = SEED,
                subset='training'
            )

validation_mask_generator = mask_data_generator.flow_from_directory(
                'data/train_mask_folder/',
                batch_size = BATCH_SIZE,
                target_size = (ORIG_HEIGHT, ORIG_WIDTH),
                seed = SEED,
                subset='validation'
            )
```

```python
number_of_train_steps = 4580 // BATCH_SIZE
number_of_val_steps = 508 // BATCH_SIZE
```

```python
print(number_of_train_steps)
print(number_of_val_steps)
```

We define Dice Loss here - this is what we will use to train the model

```python
def dice_coeff(y_true, y_pred):
    smooth = 1.
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    score = (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)
    return score

def dice_loss(y_true, y_pred):
    loss = 1 - dice_coeff(y_true, y_pred)
    return loss
```

```python
# Build U-Net model
inputs = Input((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS))

c1 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (inputs)
c1 = Dropout(0.1) (c1)
c1 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c1)
p1 = MaxPooling2D((2, 2)) (c1)

c2 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (p1)
c2 = Dropout(0.1) (c2)
c2 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c2)
p2 = MaxPooling2D((2, 2)) (c2)

c3 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (p2)
c3 = Dropout(0.2) (c3)
c3 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c3)
p3 = MaxPooling2D((2, 2)) (c3)

c4 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (p3)
c4 = Dropout(0.2) (c4)
c4 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c4)
p4 = MaxPooling2D(pool_size=(2, 2)) (c4)

c5 = Conv2D(256, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (p4)
c5 = Dropout(0.3) (c5)
c5 = Conv2D(256, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c5)

u6 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same') (c5)
u6 = concatenate([u6, c4])
c6 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (u6)
c6 = Dropout(0.2) (c6)
c6 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c6)

u7 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same') (c6)
u7 = concatenate([u7, c3])
c7 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (u7)
c7 = Dropout(0.2) (c7)
c7 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c7)

u8 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same') (c7)
u8 = concatenate([u8, c2])
c8 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (u8)
c8 = Dropout(0.1) (c8)
c8 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c8)

u9 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same') (c8)
u9 = concatenate([u9, c1], axis=3)
c9 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (u9)
c9 = Dropout(0.1) (c9)
c9 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c9)

u10 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same') (c9)
u11 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same') (u10)
crop1 = Cropping2D(cropping=((0, 0), (1, 1)))(u11)

outputs = Conv2D(1, (1, 1), activation='sigmoid') (crop1)

model = Model(inputs=[inputs], outputs=[outputs])
model.compile(optimizer='adam', loss=dice_loss, metrics=['accuracy'])
model.summary()
```

```python
my_generator_train = my_image_mask_generator(
        train_image_generator,
        train_mask_generator
    )
my_generator_val = my_image_mask_generator(
        validation_image_generator,
        validation_mask_generator
    )
```

```python
EPOCHS = 10

mc = ModelCheckpoint('outputs/weights-{epoch:03d}.h5', period=1)

model.fit_generator(
        my_generator_train,
        steps_per_epoch = number_of_train_steps,
        epochs=EPOCHS,
        validation_data=my_generator_val,
        validation_steps=number_of_val_steps,
        callbacks=[mc]
    )
```
